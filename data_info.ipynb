```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
train_dataset = pd.read_csv('train_dataset.csv')

# Sample for visualization
sample_df = train_dataset.sample(n=5000, random_state=42)

# --- Visualization 1: Count of malicious vs benign URLs ---
plt.figure(figsize=(6, 4))
sns.countplot(data=sample_df, x='label')
plt.title("Label Distribution (0 = Benign, 1 = Malicious)")
plt.xlabel("Label")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# --- Visualization 2: URL Length vs Label ---
plt.figure(figsize=(8, 5))
sns.boxplot(data=sample_df, x='label', y='url_len')
plt.title("URL Length vs Label")
plt.xlabel("Label")
plt.ylabel("URL Length")
plt.tight_layout()
plt.show()

# --- Visualization 3: Entropy vs Label ---
plt.figure(figsize=(8, 5))
sns.violinplot(data=sample_df, x='label', y='url_entropy', inner='quartile')
plt.title("URL Entropy vs Label")
plt.xlabel("Label")
plt.ylabel("Entropy")
plt.tight_layout()
plt.show()

# --- Visualization 4: Count of HTTP in URL vs Label ---
plt.figure(figsize=(8, 5))
sns.boxplot(data=sample_df, x='label', y='url_count_http')
plt.title("Count of 'http' in URL vs Label")
plt.xlabel("Label")
plt.ylabel("Count of 'http'")
plt.tight_layout()
plt.show()

# --- Visualization 5: Correlation Heatmap (Top 10 features with label) ---
corr = sample_df.corr(numeric_only=True)
top_corr = corr['label'].abs().sort_values(ascending=False).head(11)  # including label
plt.figure(figsize=(10, 8))
sns.heatmap(sample_df[top_corr.index].corr(), annot=True, cmap='coolwarm')
plt.title("Top Correlated Features with Label")
plt.tight_layout()
plt.show()
# Label Distribution

# Dataset is imbalanced: more benign URLs than malicious.

# URL Length

# Malicious URLs are longer on average — often due to obfuscation or extra parameters.

# URL Entropy

# Higher entropy (randomness) is seen in malicious URLs, indicating complex or auto-generated structures.

# 'http' Count

# Malicious URLs tend to contain 'http' multiple times, a tactic to trick users.

# Feature Correlation Heatmap

# Features like url_len, url_entropy, and url_count_sensitive_words are highly correlated with malicious labels — useful for model building.
```
